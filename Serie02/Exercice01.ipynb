### Méthode 1 : Calcul des coefficients de la droite de régression linéaire

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def regression_lineaire(x, y):
    # Calcul des moyennes
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    # Calcul des écarts à la moyenne
    x_dev = x - x_mean
    y_dev = y - y_mean

    # Calcul des coefficients de la droite de régression
    beta1 = np.sum(x_dev * y_dev) / np.sum(x_dev ** 2)
    beta0 = y_mean - beta1 * x_mean

    # Calcul de la nouvelle variable y prédite
    y_pred = beta0 + beta1 * x

    # Calcul des résidus : le résidu est la différence entre la valeur observée et la valeur prédite
    residus = y - y_pred

    # Calcul de la somme des carrés des résidus : S_err
    S_err = np.sum(residus ** 2)

    # Calcul de la somme des carrés totaux : Stot
    Stot = np.sum((y - y_mean) ** 2)

    # Calcul du coefficient de détermination : R2
    R2 = 1 - S_err / Stot

    affichage(x,y, beta1, beta0, S_err, Stot, y_pred, R2)
    return beta1, beta0, S_err, Stot, y_pred, residus, R2
```

### Méthode 2 : Affichage des résultats

```python
def affichage(x,y,beta1, beta0, S_err, Stot, y_pred, R2):
    print("Pente (beta1) : ", beta1)
    print("Ordonnée à l'origine (beta0) : ", beta0)
    print("Somme des carrés résiduels (S_err) : ", S_err)
    print("Somme totale des carrés (Stot) : ", Stot)
    print("Coefficient de détermination (R2) : ", R2)

    # Tracé de la droite de régression
    plt.scatter(x, y, color='blue')
    plt.plot(x, y_pred, color='red')
    plt.xlabel('Température (°C)')
    plt.ylabel('Rendement (%)')
    plt.show()
```

### Données d'entrée : température et rendement n=10

```python
# Données d'entrée : température et rendement n=10
x = np.array([100, 110, 120, 130, 140, 150, 160, 170, 180, 190])
y = np.array([45, 51, 54, 61, 66, 70, 74, 78, 85, 89])

# Calcul des coefficients de la droite de régression
beta1, beta0, S_err, Stot, y_pred, residus, R2 = regression_lineaire(x, y)
```

### Données d'entrée : température et rendement n=12

```python
# Données d'entrée : température et rendement n=12
x = np.array([100, 110, 120, 130, 140, 150, 160, 170, 180,190,105,180])
y = np.array([45, 51, 54, 61, 66, 70, 74, 78, 85, 89, 70, 70])

# Calcul des coefficients de la droite de régression
beta1, beta0, S_err, Stot, y_pred, residus, R2 = regression_lineaire(x, y)
```
